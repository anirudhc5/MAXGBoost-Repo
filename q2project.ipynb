```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_transaction = pd.read_csv('train_transaction.csv')
test_transaction = pd.read_csv('test_transaction.csv')
train_identity = pd.read_csv('train_identity.csv')
test_identity = pd.read_csv('test_identity.csv')

ss = pd.read_csv("sample_submission.csv")
```

```python
df_train = train_transaction.merge(train_identity, how='left', on='TransactionID')
df_test = test_transaction.merge(test_identity, how="left", on="TransactionID")
```

```python
for column in df_test.columns:
    if "id" in column: print(column); df_test.rename(columns={column: column.replace("-","_")}, inplace=True)
```

```python
print(df_train.columns.to_list())
```

```python
lst = df_train.columns.tolist()
lst.pop(df_train.columns.tolist().index("isFraud"))
lst
```

```python
missing_percent = df_train.isnull().mean() * 100
df_train = df_train.loc[:, missing_percent < 60]
df_test = df_test[lst]
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_train = df_train[df_train['TransactionAmt'] < 30000]
```

```python
df_train = df_train.drop("TransactionDT",axis=1)
```

```python
df_test = df_test.drop("TransactionDT",axis=1)
df_test = df_test.drop("TransactionID",axis=1)
```

```python
df_train = df_train.drop("TransactionID", axis=1)
```

```python
from sklearn import preprocessing
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
```

```python
X_train = df_train.drop('isFraud', axis=1)
y_train = df_train['isFraud'].copy()
X_test = df_test.copy()
```

```python
X_train.fillna(-999)
X_test.fillna(-999)
```

```python
for f in X_train.columns:
    if X_train[f].dtype=='object' or X_test[f].dtype=='object': 
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(X_train[f].values) + list(X_test[f].values))
        X_train[f] = lbl.transform(list(X_train[f].values))
        X_test[f] = lbl.transform(list(X_test[f].values))  
```

```python
clf = xgb.XGBClassifier(n_estimators=500,
                        n_jobs=4,
                        max_depth=9,
                        learning_rate=0.05,
                        subsample=0.9,
                        colsample_bytree=0.9,
                        missing=-999)

clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score
```

```python
print(classification_report(y_train, y_pred_train))
```

```python
roc_auc_score(y_train, y_pred_train)
```

```python
ss["isFraud"] = clf.predict(X_test)
```

```python
ss.to_csv('submission.csv',index=False)
```
